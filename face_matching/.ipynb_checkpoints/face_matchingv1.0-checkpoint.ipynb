# Face Matching
## Problem statement:
__'Searching Faces in Wild'.__<br>
The problem we are trying to solve here is that a user gives an image or set of images of people not seen/known in advance and the task is to search these images across a given video feed.

__Complexity:__<br>
Unlike the problem of __face recognition__ where there is an assumption of an already existing database of people on which a model can be trained(a __Supervised Learning__ approach), the problem here is a bit hard because the analysis is done on completely unknown set of people; there are not sufficient number of images per person to train a model per say(an __Unsupervised Learning__ approach).

### Using HOG
HOG features for finding facial keypoints.

#### NOTE: the code below is to show landmarks for explanation purpose on how the algorithm works internally. It has not been exposed in the API.

## HOG facial feature detection

```python
import face_recognition
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib.pyplot as plt
from skimage.feature import hog
from skimage import data, exposure
from scipy import misc
import numpy as np
from skimage import feature
from skimage import io

frame = cv2.imread('/home/anuj/Desktop/office-work/face-matching/Photos/Abhijay/Abhijay_0002.jpg')
small_frame = cv2.resize(frame, (0, 0), fx=0.25, fy=0.25)
rgb_small_frame = small_frame[:, :, ::-1]

face_locations = face_recognition.face_locations(rgb_small_frame, number_of_times_to_upsample=2, model="cnn")

# Print the location of each face in this image
(top, right, bottom, left) = face_locations[0]
top *= 4
right *= 4
bottom *= 4
left *= 4

print("A face is located at pixel location Top: {}, Left: {}, Bottom: {}, Right: {}".format(top, left, bottom, right))

# You can access the actual face itself like this:
face_image = frame[top:bottom, left:right]
# plt.axis('off')
# plt.imshow(face_image);

gray_image = cv2.cvtColor(face_image, cv2.COLOR_BGR2GRAY)
fd, hog_image = hog(gray_image, orientations=8, pixels_per_cell=(16, 16),
                cells_per_block=(1, 1), visualise=True)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5), sharex=True, sharey=True)

ax1.axis('off')
ax1.imshow(np.flip(face_image, axis=2), cmap=plt.cm.gray)
ax1.set_title('Input image')
ax1.set_adjustable('box-forced')

# Rescale histogram for better display
hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 10))
# io.imsave("hog_image.jpg", hog_image_rescaled)
ax2.axis('off')
ax2.imshow(hog_image_rescaled, cmap=plt.cm.gray)
ax2.set_title('Histogram of Oriented Gradients')
ax1.set_adjustable('box-forced')
plt.show()
```

### Using CNN 

* Face detection runs the best using CNN. 
* Using CNN with below mentioned configurable parameters helps in detecting people with NON-FRONTAL faces as well hence it would be much useful in actual scenarios. 
* We detect what is called as 'facial landmarks' on given images, these landmarks are then endcoded in a 128 dimensional, non-linear vector which has a lot more specific information about an individual face. 
* For a given test image also we compute the same feature vector and make a non-linear comparison. This is how we find faces in wild.


__Configurable parameters:__
1. number of times to upsample
2. number of jitters

**model is a ResNet network with 29 conv layers. It's essentially a version of the ResNet-34 network from the paper Deep Residual Learning for Image Recognition by He, Zhang, Ren, and Sun with a few layers removed and the number of filters per layer reduced by half.

#### NOTE: the code below is to show landmarks for explanation purpose on how the algorithm works internally. It has not been exposed in the API.

## Landmark detection using CNN

```python
import random
import numpy as np
import matplotlib.pyplot as plt
import cv2
import face_recognition

def show_landmarks(image_file, blank_image):
    """
    image_file: numpy array
    output:landmark on the image  
    """
    face_landmarks_list = face_recognition.face_landmarks(image_file)
    for landmarks in face_landmarks_list:
        for landmark_name, landmark_location in landmarks.items():
            color = (0, 255, 0)
            landmark_location = map(list,landmark_location)
            landmark_location = np.array(landmark_location)
            cv2.polylines(blank_image,[landmark_location],0,color, 2)
        
#         blank_image = np.flip(blank_image,axis=2)
    return blank_image    


input_image_1 = cv2.imread('Abhijay_0002.jpg')
blank_image = np.zeros((input_image_1.shape[0],input_image_1.shape[1],3), np.uint8)
output_image_1 = show_landmarks(input_image_1, blank_image)

# input_image_2 = cv2.imread('Nikhil_0016.jpg')
# output_image_2 = show_landmarks(input_image_2)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 10))

ax1.axis('off')
ax1.imshow(np.flip((input_image_1), axis = 2))

ax2.axis('off')
ax2.imshow(output_image_1)

plt.show()
```

## Accuracy
Accuracy has been calculated while considering some base cases below.

### Case 1. Two people being searhed, one actually present in the frame other not present
People taken: gajanan, nikhil (when gajanan is present and nikhil is not in the video) 

__Total frames__ = 200

__Miss count__ = 2

__False count__ = 7

```python
total_frames = 200
miss_count = 2.0
false_count = 7.0
accuracy = ((total_frames - (miss_count+false_count))/total_frames)*100
```

```python
print("Accuracy over {} frames is: {}%".format(total_frames, accuracy))
```

### Case 2. Person being searched is not present in the entire video
People taken: Aaron_Eckhart (when he is not present in the video) 

__Total frames__ = 200

__False count__ = 31

```python
total_frames = 200
false_count = 21.0
accuracy = ((total_frames - (false_count))/total_frames)*100
```

```python
print("Accuracy over {} frames is: {}%".format(total_frames, accuracy))
```
